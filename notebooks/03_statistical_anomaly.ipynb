```python
from data_loader import load_clean_data
import scipy

df = load_clean_data()
```

```python
mean_torque = df["Torque [Nm]"].mean()
std_torque = df["Torque [Nm]"].std()

df["torque_zscore"] = (
    df["Torque [Nm]"] - mean_torque
) / std_torque
```

```python
df["torque_anomaly"] = (
    df["torque_zscore"].abs() > 3
).astype(int)
```

```python
df["torque_anomaly"].value_counts()
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        df["Machine failure"],
        df["torque_anomaly"],
        zero_division=0
    )
)
```
